```python
import pandas as pd

# 1. Load the dataset
df = pd.read_csv('../data/aqi_dataset.csv')

# 2. Display the first 5 rows to make sure it loaded correctly
df.head()
```

```python
# 3. Check the basic info (columns, data types, and missing values)
df.info()
```

```python
# 4. Fill missing values with the median of each column
df.fillna(df.median(), inplace=True)

# 5. Check again to ensure there are exactly 0 missing values left
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 6. Plot a Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("What Impacts AQI the Most?")
plt.show()
```
